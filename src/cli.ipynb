```python
applications=[]
prep=[]
next_app_id=1
next_task_id=1

valid_status=['APPLIED','ONLINE TEST','HR SCREEN','TECH 1','TECH 2','OFFER','REJECTED']
valid_task_status=['TODO','DOING','DONE']
```

```python
#Non Empty Input
def read_non_empty(prompt,err_msg="Can't be empty.Please Try again!"):
    first=True
    while True:
        s=input(prompt if first else err_msg ).strip()
        if s:
            return s
```

```python
def read_data(prompt="Deadline (YYYY-MM-DD): "):
    s=input(prompt).strip()
    return s or " "
    
```

```python
# Application 
def add_application():
    global next_app_id
    print("\nAdd Application")
    company=read_non_empty("Company: ")
    role=read_non_empty("Role: ")
    location=input("Location(optional): ").strip()
    source=input("Source(Referal/Portal/etc): ").strip()
    deadline=read_data()
    status="APPLIED" #default
    Notes=input("Notes(optional): ").strip()

    app={
    "id":next_app_id,
    "company":company,
    "role":role,
    "location":location,
    "source":source,
    "deadline":deadline,
    "status":status,
    "notes":Notes
    }
    applications.append(app)
    print(f"Application Added #{next_app_id} ({company}-{role})")
    next_app_id+=1
```

```python
#update application statuses
def update_status():
    if not applications:
        print("\n Application not Found")
        return
    try:
        app_id=int(input("Enter Application id: "))
    except valueError:
        print("Invalid Value")
        return
    match=next((a for a in applications if a['id']==app_id),None) #if a matches then match value will be that otherwise none
    if not match:
        print("ID not found")
        return
    print("Valid Statuses: ",",".join(valid_status))
    new_status=input("New Status: ").strip().upper()
    if new_status not in valid_status:
        print("Invalid Status")
        return
    match['status']=new_status
    print(f'Updated #{app_id}->{new_status}')
    
    
```

```python
#List Applications
def list_app():
    if not applications:
        print("\n No application Added")
        return
    print("\n List Of Application ")
    filt=input("Filter by Status(Enter for All): ").strip().upper()
    rows=[a for a in applications if(not filt or a['status']==filt)]
    sort_by_deadline=input("Sort by Deadline?(y/n): ").strip().lower()=="y"
    if sort_by_deadline:
        rows.sort(key=lambda a:(a['deadline'] or '9999-99-99'))
    print("\nID  Company        Role                 Status      Deadline")
    print("--  -------------  -------------------  ----------  ----------")
    for a in rows:
            print(f'{a["id"]:>2}  {a["company"][:13]:<13}  {a["role"][:19]:<19}  {a["status"][:10]:<10}  {a["deadline"] or "-":<10}')
```

```python
def add_prep():
    global next_task_id
    topic=read_non_empty("TOPIC(e.g. Python/Sql/DSA): ").strip()
    subtopic=read_non_empty("Subtopic(e.g. Lists/Loops): ").strip()
    due_date=read_data("Due Date (YYYY-MM-DD,OPTIONAL): ").strip()
    notes=input("Notes(optional): ").strip()
    task={
    "id":next_task_id,
    "topic":topic,
    "subtopic":subtopic,
    "due_date":due_date,
    "status":"TODO",
    "notes":notes
    }
    prep.append(task)
    print(f'\n Added Task #{id} ({topic}-{subtopic})')
    next_task_id+=1
```

```python
def prep_list():
    if not prep:
        print("No task Found!")
        return
    print("\n List of Tasks")
    filt_topic=input("filter by topic: ").strip().lower()
    filt_status=input("Filter by Status(TODO/DOING/DONE): ").strip().upper()
    rows=[a for a in prep if(not filt_topic or a['topic'].lower()==filt_topic) and (not filt_status or a['status']==filt_status)]
    rows.sort(key= lambda a:(a['due_date'] or "9999-99-99"))
    print("\nID  Topic           Subtopic           Status  Due")
    print("--  ------------  -------------------  ------  ----------")
    for t in rows:
        print(f'{t["id"]:>2}  {t["topic"][:13]:<13}  {t["subtopic"][:19]:<19}  {t["status"][:6]:<6}  {t["due_date"] or "-":<10}')
        
        
    
```

```python
def update_task():
    if not prep:
        print("No task found!")
        return

    try:
        tid=int(input("Enter Id: "))
    except valueError:
        print("Invalid Id")
        return
    match=next((t for t in prep if t["id"]==tid),None)
    if not match:
        print("Invalid Id")
        return
    print("Valid Statuses ",",".join(valid_task_status))
    new_status=input("New Status: ").strip().upper()
    if new_status not in valid_task_status:
        print("Invalid Status")
        return
    match['status']=new_status
    print(f' Task Updated #{tid}->{new_status}')
```

```python
def main():
    while True:
        print("""
======== Placement Prep & Application Tracker (Milestone 1) ========
1) Add application
2) List applications
3) Update application status
4) Add prep task
5) List prep tasks
6) Update prep task 
Q) Quit
""")
        choice=input("Choose: ").strip().lower()
        if(choice=="1"):add_application()
        elif(choice=="2"):list_app()
        elif(choice=="3"):update_status()
        elif(choice=="4"):add_prep()
        elif(choice=="5"):prep_list()
        elif(choice=="6"):update_task()
        elif choice in ("q","quit"):break
        else : print("Invalid Choice")
            
        
        
```

```python
if __name__ == "__main__":
    main()
```
